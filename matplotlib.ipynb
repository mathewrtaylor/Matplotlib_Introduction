```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import openpyxl
%matplotlib inline
```

### Just The Basics

```python
#Showing a very basic example
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
# Stock Value prices in the beginning of October year over year
spce_stock = [  0.00,   10.18,    9.93,   10.38,   19.14,   23.80]
tsla_stock = [ 39.30,   71.38,   51.76,   51.39,  439.67,  775.22]
amzn_stock = [570.76, 1002.94, 1764.03, 1731.92, 3125.00, 3283.26]
year = [2016, 2017, 2018, 2019, 2020, 2021]
```

```python
# Basic Plot
plt.plot(year,spce_stock)
plt.show()
```

```python
# Plot
plt.plot(year,spce_stock)

# Now with decorations
plt.xlabel('Year')
plt.ylabel('Stock Prices')
plt.title('Symbol: SPCE')

plt.show()
```

```python
# Multiple Plots
plt.plot(year,spce_stock, '--m',label= 'SPCE')
plt.plot(year, tsla_stock, ':g',label= 'TSLA')
plt.plot(year, amzn_stock, 'b', label= 'AMZN')

# Color Options - https://matplotlib.org/stable/gallery/color/named_colors.html
# Line Styles - https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

# Decorations
plt.xlabel('Year')
plt.ylabel('Stock Prices')
plt.title('Stocks')

plt.legend()
plt.show()
```

```python
# playing with axis

# Plot
plt.plot(year,spce_stock, '--m',label= 'SPCE')
plt.plot(year, tsla_stock, ':g',label= 'TSLA')
plt.plot(year, amzn_stock, 'b', label= 'AMZN')

# Color Options - https://matplotlib.org/stable/gallery/color/named_colors.html
# Line Styles - https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

# Decorations
plt.xlabel('Year')
plt.ylabel('Stock Prices')


# Manually define axis
plt.axis([2017,2022, 10,1000])

plt.legend()
plt.show()
```

```python
#More manipulations

# Plot
plt.plot(year, spce_stock, '--m',label= 'SPCE')
plt.plot(year, tsla_stock, ':g',label= 'TSLA')
plt.plot(year, amzn_stock, 'b', label= 'AMZN')

# Decorations
plt.xlabel('Year')
plt.ylabel('Stock Prices')
# Removing axes from the figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
# Playing with the legend
plt.legend(frameon=False, loc='upper center')


plt.show()
```

### Going More in depth

```python
#Subplotting

# Rows, Columns, Position
fig_1 = plt.figure(1, figsize=(6.4, 4.8))

# Shaping
chart_1=fig_1.add_subplot(131)
chart_2=fig_1.add_subplot(132)
chart_3=fig_1.add_subplot(133)

# Plot
chart_1.plot(year,spce_stock,'m',label='SPCE')
chart_2.plot(year,tsla_stock,'g',label='TSLA')
chart_3.plot(year,amzn_stock,'b',label='AMZN')

# Decorations
fig_1.suptitle('Stock Prices')
chart_1.set_title('SPCE')
chart_2.set_title('TSLA')
chart_3.set_title('AMZN')

plt.show()
```

```python
# Flipping the script!
fig_1 = plt.figure(1, figsize=(6.4, 4.8))

# Shaping
chart_1=fig_1.add_subplot(311)
chart_2=fig_1.add_subplot(312)
chart_3=fig_1.add_subplot(313)

# Plot
chart_1.plot(year,spce_stock,'m',label='SPCE')
chart_2.plot(year,tsla_stock,'g',label='TSLA')
chart_3.plot(year,amzn_stock,'b',label='AMZN')

# Decorations
fig_1.suptitle('Stock Prices')
chart_1.set_title('SPCE')
chart_2.set_title('TSLA')
chart_3.set_title('AMZN')

plt.show()
```

```python
# Adjusting Figure size
fig_1 = plt.figure(1, figsize=(10, 5))

# Shaping
chart_1=fig_1.add_subplot(311)
chart_2=fig_1.add_subplot(312)
chart_3=fig_1.add_subplot(313)

# Plot
chart_1.plot(year,spce_stock,'m',label='SPCE')
chart_2.plot(year,tsla_stock,'g',label='TSLA')
chart_3.plot(year,amzn_stock,'b',label='AMZN')

# Decorations
fig_1.suptitle('Stock Prices')
chart_1.set_title('SPCE')
chart_2.set_title('TSLA')
chart_3.set_title('AMZN')

plt.show()
```

```python
# Alternate way to add axes, and axes != axis
fig_2, axes = plt.subplots(3,1,figsize=(10,5))
plt.show()
```

```python
fig_2, axes = plt.subplots(3,1,figsize=(10,5), sharex=True)

# Plot
fig_2.axes[0].plot(year,spce_stock,'m')
fig_2.axes[1].plot(year,tsla_stock,'g')
fig_2.axes[2].plot(year,amzn_stock,'b')

# Decorations
fig_2.suptitle('Stock Prices')
fig_2.axes[0].set_title('SPCE')
fig_2.axes[1].set_title('TSLA')
fig_2.axes[2].set_title('AMZN')
plt.show()
```

```python
# So what's possible?
fig_3, axes = plt.subplots(3,3,figsize=(20,10), sharex=True)

# Plot
fig_3.axes[0].plot(year,spce_stock,'m')
fig_3.axes[1].plot(year,tsla_stock,'g')
fig_3.axes[2].plot(year,amzn_stock,'b')
fig_3.axes[3].bar(year,spce_stock)
fig_3.axes[4].bar(year,tsla_stock)
fig_3.axes[5].bar(year,amzn_stock)
fig_3.axes[6].scatter(year,spce_stock)
fig_3.axes[7].scatter(year,tsla_stock)
fig_3.axes[8].scatter(year,amzn_stock)

# Decorations
fig_3.axes[0].set_title('SPCE')
fig_3.axes[1].set_title('TSLA')
fig_3.axes[2].set_title('AMZN')
fig_3.suptitle('Stock Prices')

fig_3.axes[0].text(2021, 23.80, 'Launch')
fig_3.axes[4].annotate('Launch', xy=(2021, 775.22), xytext=(2020, 600),
            arrowprops=dict(facecolor='black', shrink=0.05))
fig_3.axes[8].text(2021, 2800, 'Launch', ha='right', va='center', rotation=60, size=10,
    bbox=dict(boxstyle='rarrow,pad=0.3', fc='cyan', ec='b', lw=2))
# https://matplotlib.org/stable/tutorials/text/annotations.html#annotations-tutorial

plt.show()
```

```python
#Even more manipulations
fig_4, axes = plt.subplots()

# Plot
fig_4.axes[0].plot(year, spce_stock, '--m',label= 'SPCE')
fig_4.axes[0].plot(year, tsla_stock, ':g',label= 'TSLA')
fig_4.axes[0].plot(year, amzn_stock, 'b', label= 'AMZN')

# Decorations
plt.xlabel('Year')
plt.ylabel('Stock Prices')
fig_4.suptitle('Stock Prices')
# Removing axes from the figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Placing a watermark
# https://matplotlib.org/stable/gallery/text_labels_and_annotations/watermark_text.html
fig_4.text(0.25, 0.35, 'Made by: Mathew', transform=axes.transAxes,
        fontsize='medium', color='lightgray', 
        ha='center', va='center', rotation='20')
# Adding some annotations
fig_4.axes[0].text(2019.75, 3100, 'Blue Origins', ha='right', va='center', rotation=0, size=10,
    bbox=dict(boxstyle='rarrow,pad=0.3', fc='cyan', ec='b', lw=2))
fig_4.axes[0].text(2019., 52, 'Tesla in Space', ha='right', va='center', rotation=5, size=10,
    bbox=dict(boxstyle='rarrow,pad=0.3', fc='g', ec='g', lw=2))
fig_4.axes[0].text(2016, 220, 'VSS Unity', ha='right', va='center', rotation=-18, size=10,
    bbox=dict(boxstyle='rarrow,pad=0.3', fc='m', ec='m', lw=2))

# Playing with the legend
fig_4.legend(frameon=False, loc='center right')

plt.show()
```

### Getting Fancier

```python
# https://www.machinelearningplus.com/plots/top-50-matplotlib-visualizations-the-master-plots-python/ 

import pandas as pd

# Prepare Data
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
x = df.loc[:, ['mpg']]
df['mpg_z'] = (x - x.mean())/x.std()
df['colors'] = ['red' if x < 0 else 'green' for x in df['mpg_z']]
df.sort_values('mpg_z', inplace=True)
df.reset_index(inplace=True)

# Draw plot
plt.figure(figsize=(14,10), dpi= 80)
plt.hlines(y=df.index, xmin=0, xmax=df.mpg_z, color=df.colors, alpha=0.4, linewidth=5)

# Decorations
plt.gca().set(ylabel='$Model$', xlabel='$Mileage$')
plt.yticks(df.index, df.cars, fontsize=12)
plt.title('Diverging Bars of Car Mileage', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

### Introducing Seaborn! <br>
Source: https://seaborn.pydata.org/

```python
#some Examples

import seaborn as sns
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
f.suptitle('Annotated heatmaps')
plt.show()
```

```python
#Interesting Hex Plot

sns.set_theme(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

p = sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
p.fig.suptitle("Hexbin plot with marginal distributions")
plt.show()
```

```python
#Like a box plot, but cooler!
sns.set_theme()

# Create a random dataset across several variables
rs = np.random.default_rng(0)
n, p = 40, 8
d = rs.normal(0, 2, (n, p))
d += np.log(np.arange(1, p + 1)) * -5 + 10

# Show each distribution with both violins and points
fig_5, ax = plt.subplots(figsize=(9, 6))
fig_5 = sns.violinplot(data=d, palette="light:r", inner="points", orient="h")
fig_5.set_title("Violinplots with observations")

plt.show()
```

```python
# Saving to an image
fig_4.savefig('Space_Stocks.png', format='png')

# Creating the DataFrame and Saving it to an Excel Sheet
data = [year,spce_stock,tsla_stock,amzn_stock]
column_names = data.pop(0)
df = pd.DataFrame(data, columns=column_names)
df.index=['SPCE','TSLA','AMZN']

with pd.ExcelWriter('python_plot.xlsx',engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Stocks', index=False)

# Placing the chart below that chart
wb = openpyxl.load_workbook('python_plot.xlsx')
ws = wb.active
my_png = openpyxl.drawing.image.Image('Space_Stocks.png')
ws.add_image(my_png, 'B6')
wb.save('python_plot.xlsx')
```
